# **export the output to diffrent layers dxf&images**

```python

import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from dfp.utils.rgb_ind_convertor import floorplan_room_map, floorplan_boundary_map
!pip install ezdxf
import cv2
import numpy as np
import ezdxf
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python

def extract_polylines(mask):
    contours, _ = cv2.findContours(mask.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    polylines = []
    for contour in contours:
        epsilon = 0.01 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)
        polylines.append(approx)
    return polylines

def export_polylines_to_dxf(room_polylines, wall_polylines, window_door_polylines, wallcircu_polylines, filename):
    doc = ezdxf.new('R2010')
    msp = doc.modelspace()

    # Create layers and set colors
    doc.layers.new(name='Rooms', dxfattribs={'color': ezdxf.rgb2int(floorplan_room_map[3])})  # livingroom/kitchen/diningroom color
    doc.layers.new(name='Walls', dxfattribs={'color': ezdxf.rgb2int(floorplan_boundary_map[2])})  # wall line color
    doc.layers.new(name='Windows_Doors', dxfattribs={'color': ezdxf.rgb2int(floorplan_boundary_map[1])})  # opening (door&window) color
    doc.layers.new(name='Wall_Circumference', dxfattribs={'color': ezdxf.rgb2int(floorplan_fuse_map[10])})  # wall circumference color

    # Add room polylines
    for polyline in room_polylines:
        points = [tuple(point[0]) for point in polyline]
        msp.add_lwpolyline(points, close=True, dxfattribs={'layer': 'Rooms'})

    # Add wall polylines
    for polyline in wall_polylines:
        points = [tuple(point[0]) for point in polyline]
        msp.add_lwpolyline(points, close=True, dxfattribs={'layer': 'Walls'})

    # Add window & door polylines
    for polyline in window_door_polylines:
        points = [tuple(point[0]) for point in polyline]
        msp.add_lwpolyline(points, close=True, dxfattribs={'layer': 'Windows_Doors'})

    # Add wall circumference polylines
    for polyline in wallcircu_polylines:
        points = [tuple(point[0]) for point in polyline]
        msp.add_lwpolyline(points, close=True, dxfattribs={'layer': 'Wall_Circumference'})

    doc.saveas(filename)

# Extract polylines for room segmentation
room_mask = (newr > 0).astype(np.uint8)
room_polylines = extract_polylines(room_mask)

# Extract polylines for wall segmentation
wall_mask = (newcw == 2).astype(np.uint8)
wall_polylines = extract_polylines(wall_mask)

# Extract polylines for windows & door segmentation
window_door_mask = (newcw == 1).astype(np.uint8)
window_door_polylines = extract_polylines(window_door_mask)

# Extract polylines for wall circumference
wallcircu_mask = (newcw > 0).astype(np.uint8)
wallcircu_polylines = extract_polylines(wallcircu_mask)

# Create images with polylines
original_img = mpimg.imread(args.image)[:, :, :3]
room_polyline_img = original_img.copy()
wall_polyline_img = original_img.copy()
window_door_polyline_img = original_img.copy()
wallcircu_polyline_img = original_img.copy()
combined_polyline_img = original_img.copy()

# Draw room polylines
cv2.polylines(room_polyline_img, room_polylines, True, (255, 0, 0), 2)

# Draw wall polylines
cv2.polylines(wall_polyline_img, wall_polylines, True, (0, 255, 0), 2)

# Draw windows & door polylines
cv2.polylines(window_door_polyline_img, window_door_polylines, True, (0, 0, 255), 2)

# Draw wall circumference polylines
cv2.polylines(wallcircu_polyline_img, wallcircu_polylines, True, (255, 255, 0), 2)

# Draw all polylines on the combined image
cv2.polylines(combined_polyline_img, room_polylines, True, (255, 0, 0), 2)
cv2.polylines(combined_polyline_img, wall_polylines, True, (0, 255, 0), 2)
cv2.polylines(combined_polyline_img, window_door_polylines, True, (0, 0, 255), 2)
cv2.polylines(combined_polyline_img, wallcircu_polylines, True, (255, 255, 0), 2)

# Display the results
plt.figure(figsize=(20, 20))

plt.subplot(2, 3, 1)
plt.imshow(original_img)
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(room_polyline_img)
plt.title("Room Segmentation Polylines")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(wall_polyline_img)
plt.title("Wall Segmentation Polylines")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(window_door_polyline_img)
plt.title("Windows & Door Segmentation Polylines")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(wallcircu_polyline_img)
plt.title("Wall Circumference Polylines")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(combined_polyline_img)
plt.title("Combined Polylines")
plt.axis('off')

plt.tight_layout()
plt.show()

# Save the results
output_dir = "./polyline_results"
os.makedirs(output_dir, exist_ok=True)

mpimg.imsave(os.path.join(output_dir, "room_polylines.png"), room_polyline_img)
mpimg.imsave(os.path.join(output_dir, "wall_polylines.png"), wall_polyline_img)
mpimg.imsave(os.path.join(output_dir, "window_door_polylines.png"), window_door_polyline_img)
mpimg.imsave(os.path.join(output_dir, "wall_circumference_polylines.png"), wallcircu_polyline_img)
mpimg.imsave(os.path.join(output_dir, "combined_polylines.png"), combined_polyline_img)

# Export polylines to DXF
dxf_filename = os.path.join(output_dir, "floorplan_polylines.dxf")
export_polylines_to_dxf(room_polylines, wall_polylines, window_door_polylines, wallcircu_polylines, dxf_filename)

print(f"Polyline images and DXF file saved in {output_dir}")
```
